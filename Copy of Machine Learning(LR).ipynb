```python
#importing libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
#from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
#op_data = pd.read_csv('sample_submission.csv')
```

```python
features = ['deg_C', 'relative_humidity', 'absolute_humidity', 'sensor_1', 'sensor_2', 'sensor_3', 'sensor_4', 'sensor_5']
target = ['target_carbon_monoxide', 'target_benzene', 'target_nitrogen_oxides']
```

```python
X_train, X_val, y_train, y_val = train_test_split(train_data[features], train_data[target], test_size=0.2, random_state=42)
```

```python
lr_co = LinearRegression()
lr_co.fit(X_train, y_train['target_carbon_monoxide'])
co_pred = lr_co.predict(X_val)

lr_bz = LinearRegression()
lr_bz.fit(X_train, y_train['target_benzene'])
bz_pred = lr_bz.predict(X_val)

lr_no = LinearRegression()
lr_no.fit(X_train, y_train['target_nitrogen_oxides'])
no_pred = lr_no.predict(X_val)
```

```python
#from sklearn.metrics import mean_squared_error

mse_carbon_monoxide = mean_squared_error(y_val['target_carbon_monoxide'], co_pred)
mse_benzene = mean_squared_error(y_val['target_benzene'], bz_pred)
mse_nitrogen_oxides =mean_squared_error(y_val['target_nitrogen_oxides'], no_pred)

print('MSE for carbon monoxide:', mse_carbon_monoxide)
print('MSE for benzene:', mse_benzene)
print('MSE for nitrogen oxides:', mse_nitrogen_oxides)
```

```python
carbon_monoxide_test_preds = lr_co.predict(test_data[features])
benzene_test_preds = lr_bz.predict(test_data[features])
nitrogen_oxides_test_preds = lr_no.predict(test_data[features])

submission = pd.DataFrame({
    'date_time': test_data['date_time'],
    'target_carbon_monoxide': carbon_monoxide_test_preds,
    'target_benzene': benzene_test_preds,
    'target_nitrogen_oxides': nitrogen_oxides_test_preds
})

submission.to_csv('submission.csv', index=False)
```
